```python
""" Write a program to count the number of each characters in a string """
try:
  str = input("Enter a String:")
  if not str:
    raise ValueError("Expecting something!")
  print("The count of each letters in the string:",str)
  new_str = {}
  for i in str:
    new_str[i] = str.count(i)  
  print(new_str) 
except ValueError as e:
  print(e)
```

Output:
Enter a String:hello world
The count of each letters in the string: hello world
{'h': 1, 'e': 1, 'l': 3, 'o': 2, ' ': 1, 'w': 1, 'r': 1, 'd': 1}
