```python
""" Complete the following code to find the minimum of 3 number using conditional statements. Output should be as displayed """
try:
  a,b,c = input("Enter three numbers followed by  : ").split(":")
  print("First number :",int(a))
  print("Second number :",int(b))
  print("Third number :",int(c))
  if(a==b and a==c):
      print("Entered numbers are equal!!!")
  elif(a<c and a<b):
      print(a," is smallest")
  elif(b<a and b<c):
      print(b," is smallest")
  else:
      print(c," is smallest")
except:
  print("Expecting a numeric value!")
```

Output:
Enter three numbers followed by  : 1:2:3
First number : 1
Second number : 2
Third number : 3
1  is smallest
