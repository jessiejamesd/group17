```python
""" Write a program to find the area and perimeter of a rectangle using functions """
def rect_area(l, b):
  return l*b
def rect_peri(l, b):
  return 2*(l+b)
try:
  l,b = input("Enter the length & breadth separated by comma").split(",")
  print("Area with length",l,"and breadth",b,"is",rect_area(int(l),int(b)))
  print("Perimeter with length",l,"and breadth",b,"is",rect_peri(int(l),int(b)))
except:
  print("Expecting a numeric value!")
```

Area with length 2 and breadth 3 is 6
Perimeter with length 2 and breadth 3 is 10
