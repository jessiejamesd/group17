```python
""" Complete the following code to convert hour into seconds. Output should be as displayed """
def to_seconds(t):     
  return t*3600
try:
  time_in_hours = int(input("Enter the time in hours:"))
  print(time_in_hours ," Hour is equal to" ,to_seconds(time_in_hours) ," Seconds")
except:
  print("Expecting a numeric value")
```

Output:
Enter the time in hours:10
10  Hour is equal to 36000  Seconds
